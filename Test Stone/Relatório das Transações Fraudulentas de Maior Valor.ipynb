```python
import pandas as pd
```

```python
frauds = pd.read_csv('data/dataset-frauds.csv', sep = ';')
```

```python
frauds
```

```python
transactions = pd.read_csv('data/dataset-transactions.csv', sep = ';')
```

```python
transactions
```

```python
transactions_selector = transactions['id'].isin(frauds['transaction_id'])
```

```python
transactions_selector
```

```python
fraud_trasactions = transactions[transactions_selector]
```

```python
fraud_trasactions = fraud_trasactions.sort_values(by = 'value', ascending = False)
```

```python
fraud_trasactions.index = range(fraud_trasactions.shape[0])
```

```python
fraud_trasactions
```

```python
print(' A maior transação fraudulenta, possui o id = {} e o valor de = {}R$'.format(fraud_trasactions['id'].iloc[0],
                                                                                 fraud_trasactions['value'].iloc[0]))
```
